```python
from PIL import Image
from torchvision import transforms

from carla_drl.autoencoder.model import VariationalAutoencoder
```

```python
ss_ae_model_path = "path/to/ss_ae_model.pth"
ss_ae_model = VariationalAutoencoder(3, 50)
ss_ae_model.load(ss_ae_model_path)
```

```python
ss_image = Image.open("../data/semseg/Town02_Opt/ClearNoon/000.png").convert("RGB")
```

```python
transform = transforms.Compose([
    transforms.Resize((80, 160)),
    transforms.ToTensor(),
])
```

```python
ss_image_tensor = transform(ss_image)
```

```python
ss_reconstructed = ss_ae_model.decoder(ss_ae_model.encoder(ss_image_tensor.unsqueeze(0)))
ss_pred_im = transforms.functional.to_pil_image(ss_reconstructed[0])
```
